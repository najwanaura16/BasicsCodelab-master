```python
# Langkah 1: Import Library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Langkah 2: Load Dataset (unggah file tips.csv di Google Colab sebelum menjalankan ini)
df = pd.read_csv('tips.csv')
```

```python
# Langkah 3: Hitung jumlah berdasarkan jenis kelamin
sex_counts = df['sex'].value_counts()
sex_percent = df['sex'].value_counts(normalize=True) * 100
```

```python
# Langkah 4: Pie Chart Visualisasi
plt.figure(figsize=(6, 6))
colors = ['skyblue', 'lightcoral']
plt.pie(sex_percent, labels=sex_percent.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Gender Pemberi Tips')
plt.axis('equal')  # agar pie chart berbentuk lingkaran sempurna
plt.show()
```
